```python
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait as wait
from selenium.webdriver.support import expected_conditions as EC
```

```python
def get_driver():
    options = webdriver.ChromeOptions()
#     options.add_argument('--headless')
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    service = Service(executable_path=r"C:\\Users\\kunal\\Desktop\\Parth\\Assignment\\chromedriver.exe")
    driver = webdriver.Chrome(service=service,options=options)
    return driver
```

```python
imdb_url = "https://www.imdb.com/search/name/"
wiki_url = "https://www.wikipedia.org/"
actor_name = 'sanjay mishra (actor)'
```

```python
def imdb_search_actor_page(actor_name):
    driver = get_driver()
    driver.get(imdb_url)
    search_box = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[2]/div[3]/form/div/div[1]/div[2]/input')
    search_box.send_keys(actor_name)
    search_button = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[2]/div[3]/form/div/p[3]/button')
    search_button.send_keys(Keys.ENTER)
    try:
        actor = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[2]/div[3]/div[1]/div/div[3]/div[1]/div[2]/h3/a') 
    except NoSuchElementException:
        logging.error('No such actor exists!')
        return None
    actor.send_keys(Keys.ENTER)
#     print(driver.find_element(By.XPATH, '/html/body/div[2]/main/div/section[1]/div/section/div/div[1]/div[4]').text)
    table = wait(driver, 60).until(EC.presence_of_element_located((By.XPATH, '/html/body/div[2]/main/div/section[1]/div/section/div/div[1]/div[4]/section[2]')))
    
def extract_movie_list(driver):
    i = 1
    movie_list = []
    try:
        while 1:
            movie_name = driver.find_element(By.XPATH, f'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[3]/tbody/tr[{i}]/td[2]').text
            movie_list.append(movie_name)
            i+=1
    except NoSuchElementException:
        return movie_list

def wiki_search_actor_page(actor_name):
    driver = get_driver()
    movie_list = []
    driver.get(wiki_url)
    search_box = driver.find_element(By.XPATH, '/html/body/div[3]/form/fieldset/div/input')
    search_box.send_keys(actor_name)
    driver.find_element(By.XPATH, '/html/body/div[3]/form/fieldset/button').send_keys(Keys.ENTER)
    movie_list = extract_movie_list(driver)
    return movie_list
```
